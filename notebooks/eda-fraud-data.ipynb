```python


print(f"Dataset Shape: {df_credit.shape}")
print(f"Class Distribution:\n{df_credit['Class'].value_counts(normalize=True) * 100}")

# 1. Statistical Summary of the Transaction Amount for Fraud vs Legitimate
print("\n--- Transaction Amount Statistics by Class ---")
print(df_credit.groupby('Class')['Amount'].describe())

# 2. Visualizing Transaction Volume Distribution (Violin Plot)
plt.figure(figsize=(8, 5))
sns.violinplot(x='Class', y='Amount', data=df_credit[df_credit['Amount'] < 500], palette='Set1')
plt.title('Distribution of Transaction Amounts (< $500) by Class')
plt.xlabel('Class (0: Legitimate, 1: Fraud)')
plt.ylabel('Transaction Amount ($)')
plt.savefig('credit_amount_distribution.png')
plt.show()

# 3. Correlation Matrix of PCA Features with the Target Class
# Since there are 28 V-features, let's find the top 8 features most correlated with Fraud
correlations = df_credit.corr()['Class'].sort_values()
print("\nTop Negative Correlations with Fraud:\n", correlations.head(4))
print("\nTop Positive Correlations with Fraud:\n", correlations.tail(5))

# Plotting Top Discriminating PCA Components
top_features = ['V17', 'V14', 'V12', 'V11', 'V4', 'Class']
plt.figure(figsize=(10, 8))
sns.heatmap(df_credit[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Top Discriminating PCA Features with Class')
plt.savefig('credit_pca_correlation.png')
plt.show()
```
